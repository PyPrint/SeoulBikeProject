## Part 1: Exploring the data and preprocessing.

First we import the libraries needed.

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import power_transform
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, accuracy_score

import tensorflow as tf
```

We manually download the dataset, and load it using Pandas. We then explore the data a bit to see if there are any inconsistencies or features/values that we need to be careful with.

```python
df = pd.read_csv('SeoulBikeData.csv', encoding= 'unicode_escape')
df.tail()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Ideally, we want that numerical features all fall within a certain interval (0,1) or (-3,3) and we want that categorical features obtain values in a small number of classes. However, if we would convert the date column to a numerical column, we would obtain 365 classes. If we apply a one-hot enconding to this, we would obtain vectors with length 365. This is not what we want, so instead we convert the dates to weekdays. 

```python
df['Weekday'] = [datetime.datetime.strptime(date, '%d/%m/%Y').weekday() for date in df['Date']] #monday = 0, sunday = 6
```

We convert the categorical columns: Weekday, Hour, Seasons, Holiday and Functioning Day to dummy variables. 

```python
df_processed = pd.get_dummies(data=df, columns=['Weekday', 'Hour', 'Seasons', 'Holiday', 'Functioning Day'])
df_processed.head()
```

We will drop the Date column for trivial reasons. Also we will check if the variance is high. If it is, we will scale the data appropriately. 

```python
df_processed.drop(['Date'], axis=1, inplace=True)
list(zip(list(df_processed.std()), list(df_processed.columns)))
```

We see that our target variable, Rented Bike Count, has a high variance, as well as the Visibility variable. Let us see what the distributions of these variables are. 

```python
plt.hist(x=df_processed['Rented Bike Count'])
plt.xlabel('Rented Bike Count per hour')
plt.ylabel('Frequency')
plt.show()
```

We do not want to use the minmax scaler for this, because our data does not have a normal distribution. Also, we will not use the normalization scaler, because this will just move the data to the left and shrinks it. The log transform is a very good option, but we want to use the BoxCox scaler, because all data is positive. We do need to be careful with zero. 

```python
df_processed['Rented Bike Count'] = power_transform(np.array(df_processed['Rented Bike Count'].apply(lambda x : x+1)).reshape(-1, 1), method='box-cox') 
df_processed.head()
```

```python
plt.hist(x=df_processed['Visibility (10m)'])
plt.xlabel('Visibility (10m)')
plt.ylabel('Frequency')
plt.show()
```

```python
df_processed['Visibility (10m)'] = power_transform(np.array(df_processed['Visibility (10m)']).reshape(-1, 1), method='box-cox') 
df_processed.head()
```

Let us check if the variance problem is resolved..

```python
list(zip(list(df_processed.std()), list(df_processed.columns)))
```

That is better!

## Part 2: Using Scikit-Learn and Keras to predict the number of bikes that need to be available. 

Let us define the feature values X and the target values y. And using the train_test_split function, we define the training set and the test set. 

```python
X = df_processed.drop('Rented Bike Count', axis=1)
y = df_processed['Rented Bike Count']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
LR = LinearRegression().fit(X_train, y_train)
predictions = LR.predict(X_test)
```

```python
list(zip(list(predictions), list(y_test)))
```

```python
LinearRegression_MAE = mean_absolute_error(predictions, y_test)
LinearRegression_MSE = mean_squared_error(predictions, y_test)
LinearRegression_R2 = r2_score(predictions, y_test)
Report = pd.DataFrame({'MAE': LinearRegression_MAE, 'MSE' : LinearRegression_MSE, 'R2' : LinearRegression_R2}, index=[0])
Report
```

We want the R2 score close to 1. Now, lets design a simple deep neural network using the Keras API.

The input shape is (48,) because we have 48 columns. We want two hidden layers of 16 neurons with Leaky Relu activation, the output layer will have linear output. 

```python
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(48,)))
model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.LeakyReLU(alpha=0.3))
model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.LeakyReLU(alpha=0.3))
model.add(tf.keras.layers.Dense(1, activation='linear'))

model.summary()
```

```python
optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3)
loss=tf.keras.losses.MeanSquaredError()
model.compile(optimizer=optimizer, loss=loss)
# This builds the model for the first time:
model.fit(X_train, y_train, batch_size=32, epochs=10)
```

```python
predictions = model.predict(X_test)
list(zip(list(predictions), list(y_test)))
```
